```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.preprocessing import LabelEncoder
```

## Domain – Banking/Loan 
### focus –  Lower NPA (Non Performing Asset) 

```python
data = pd.read_csv(r'C:\Users\nimesh.yadav\Desktop\Python\Module_7\Datasets\loan_borowwer_data.csv')
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.describe()
```

```python
purpose_values = np.array(data['purpose'])
encoder = LabelEncoder()
encoded_values = encoder.fit_transform(purpose_values)
data['purpose'] = encoded_values
```

```python
data.head()
```

```python
corr = data.corr()
plt.figure(figsize = (8,6))
sns.heatmap(corr.round(1),annot=True)
plt.show()
```

```python
sns.countplot(data['not.fully.paid'])
```

```python
Y = data.pop('not.fully.paid')
X = data
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.3 , random_state = 50)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
Y_train.shape
```

```python
Y_test.shape
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
Y_pred = model.predict(X_test)
```

#### Accuracy of the model is :

```python
Accuracy = metrics.accuracy_score(Y_test,Y_pred)
print(f"Accuracy of the model is : {(Accuracy*100).round(2)} %.")
```

```python
result_df = pd.DataFrame({'Actual_output': Y_test, 'Predicted_output' : Y_pred})
result_df
```

```python
metrics.confusion_matrix(Y_test,Y_pred)
```
