### 2. This dataset contains many categorical features, replace them with label encoding and 
### 3. Replace the missing values by the most frequent value in each column. 

```python
all_col = df.columns
non_cat_col = df.describe().columns
```

```python
def Diff(li1,li2):
    return (list(set(li1)-set(li2)))
```

```python
cat_col_list = Diff(all_col,non_cat_col)
cat_col_list
```

```python
df[cat_col_list].head()
```

```python
df[cat_col_list].info()
```

```python
def fillna(col):
    df.fillna(col.mode().iloc[0], inplace=True)
    return col
```

```python
df = df.apply(lambda x:x.fillna(x.value_counts().index[0]))
```

```python
pd.isna(df).head(10)
```

```python
encoder = LabelEncoder()
for i in cat_col_list:
    df[i] = encoder.fit_transform(np.array(df[i]))
```

```python
df.columns
```

```python
df.head()
```

### 4. Fit a decision tree classifier and observe the accuracy.

```python
sns.countplot(df['outcome'])
plt.show()
```

```python
Y = df['outcome']
X = df.drop('outcome',axis=1)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.3 , random_state = 85)
```

```python
X_train.head()
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
Y_train.shape
```

```python
Y_test.shape
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_train,Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
metrics.accuracy_score(Y_test,Y_pred)
```

### 5. Fit a random forest classifier and observe the accuracy.

```python
from sklearn.ensemble import RandomForestClassifier

random_model = RandomForestClassifier(n_estimators=100)

# training
random_model.fit(X_train, Y_train)

# testing
Y_pred = random_model.predict(X_test)

# accoracy
metrics.accuracy_score(Y_test, Y_pred)
```
