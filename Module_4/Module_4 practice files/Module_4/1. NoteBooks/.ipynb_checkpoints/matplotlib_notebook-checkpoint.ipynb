# Data Visualization using matplotlib

### Line plot

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 40])  # when one value is passed then i.e y-axis value
plt.show()  # displays the plot
```

```python
# line plot by providing both x and y values
x = [1, 2, 3, 4]
y = [elem**2 for elem in x]
plt.plot(x,y)
plt.show()
```

```python
# line plot by providing both x and y values
import numpy as np
x = np.arange(0,5,0.01)
y = [elem**2 for elem in x]
plt.plot(x,y)
plt.show()
```

```python
# line plot by providing both x and y values
import numpy as np
x = [10, 20, 30, 40]
y = [1, 2, 3, 4]
plt.plot(x,y)
plt.show()
```

# Multiline plot
means, plotting multiple plots together

```python
# method 1: calling .plot() for each pair of x-y
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
plt.plot(x, [elem for elem in x])
plt.plot(x, [elem**2 for elem in x])
plt.plot(x, [elem**3 for elem in x])
plt.show()
```

```python
# method 2: calling .plot() only one time
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
plt.plot(x, [elem for elem in x],
         x, [elem**2 for elem in x],
         x, [elem**3 for elem in x])
plt.show()
```

### adding a grid

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
plt.plot(x, [elem for elem in x],
         x, [elem**2 for elem in x],
         x, [elem**3 for elem in x])
plt.grid(True)  # add the grid
plt.show()
```

### Limiting the axes
providing the axis range

```python
# Method 1: using .axis() method
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
plt.plot(x, [elem for elem in x],
         x, [elem**2 for elem in x],
         x, [elem**3 for elem in x])
plt.grid(True)  # add the grid
plt.axis([-1, 2, -1, 10])  # (x_axis_start, x_axis_end, y_axis_start, y_axis_end)
plt.show()
```

```python
# Method 2: using xlim() and ylim() method
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
plt.plot(x, [elem for elem in x],
         x, [elem**2 for elem in x],
         x, [elem**3 for elem in x])
plt.grid(True)  # add the grid
#plt.axis([-1, 2, -1, 10])  # (x_axis_start, x_axis_end, y_axis_start, y_axis_end)
plt.xlim(-1, 2)
plt.ylim(-1, 10)
plt.show()
```

### adding labels
adding names to x and y axis

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
plt.plot(x, [elem for elem in x],
         x, [elem**2 for elem in x],
         x, [elem**3 for elem in x])
plt.grid(True)  # add the grid
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

### adding a title for the graph

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
plt.plot(x, [elem for elem in x],
         x, [elem**2 for elem in x],
         x, [elem**3 for elem in x])
plt.grid(True)  # add the grid
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('My Plot')
plt.show()
```

### adding a Legend
giving name to each plot

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
plt.plot(x, [elem for elem in x], label='linear')
plt.plot(x, [elem**2 for elem in x], label='square')
plt.plot(x, [elem**3 for elem in x], label='cube')
plt.grid(True)  # add the grid
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()  # assign labels to the plot
plt.show()
```

### Saving Plots
using savefig()

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
plt.plot(x, [elem for elem in x], label='linear')
plt.plot(x, [elem**2 for elem in x], label='square')
plt.plot(x, [elem**3 for elem in x], label='cube')
plt.grid(True)  # add the grid
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()  # assign labels to the plot

plt.savefig('myplot.png') # saves the plot at the location mentioned in the paramterer

plt.show()
```

## Bar chart

```python
import matplotlib.pyplot as plt
plt.bar([1, 2, 3], [1, 4, 9])
plt.show()
```

### plot a dictionary using Bar chart

```python
my_dict = {'A':20, 'B':65, 'C': 50, 'D': 80}
for i, key in enumerate(my_dict):
    plt.bar(i, my_dict[key])
plt.show()
```

```python
# giving name to each bar
my_dict = {'A':20, 'B':65, 'C': 50, 'D': 80}
for i, key in enumerate(my_dict):
    plt.bar(i, my_dict[key])

plt.xticks([0, 1, 2, 3],  my_dict.keys())
plt.show()
```

## Pie chart
used to compare the multiple parts against the whole

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5,5)) # size of the plot
x = [60, 20, 10, 7, 3] # proportion of sectors
my_labels = ['Bikes', 'Cars', 'Super-cars', 'bicyle', 'planes']
plt.pie(x, labels=my_labels)
plt.show()

```

## show percentage in pie-plot
using the parameter 'autopct'

```python
import matplotlib.pyplot as plt

names = ['aa', 'bb', 'cc', 'dd']

fig1, ax1 = plt.subplots()
ax1.pie([30, 25, 40, 35],
        labels=names,
        autopct='%1.1f%%'
)

ax1.axis('equal')
plt.show()
```

```python
# using explode(): to separate a part of a pie chart

import matplotlib.pyplot as plt

names = ['aa', 'bb', 'cc', 'dd']

fig1, ax1 = plt.subplots()
ax1.pie([30, 25, 40, 35],
        explode=(0, 0.1, 0, 0), # it would separate 'bb'
        labels=names,
        autopct='%1.1f%%'
)

ax1.axis('equal')
plt.show()

```

# Styling

```python
# choosing colours

import matplotlib.pyplot as plt
import numpy as np

y = np.arange(1, 5)
print(y)
plt.plot(y, 'y')
plt.plot(y+10, 'm')
plt.plot(y+20, 'c')
plt.show()
```

```python
# control line styling
import matplotlib.pyplot as plt
import numpy as np

y = np.arange(1, 100)

# specifying line styling : --(solid line), -.(solid-dash) & : (dotted)
plt.plot(y, '--', y*5, '-.', y*10, ':')

plt.show()

```

#### control markers styling

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.arange(1, 3, 0.2)

plt.plot(y, '*', y+0.5, 'o', y+1, 'D', y+2, '^', y+3, 's')

plt.show()

```
