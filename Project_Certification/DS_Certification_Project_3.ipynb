## Encoding Target string data using LabelEncoder

```python
def encoder(data):
    le = LabelEncoder().fit(data)
    encoded_data = le.transform(data)
    classes = le.classes_
    return encoded_data,classes,le   
```

```python
Y,classes,encod_model = encoder(Y)
```

```python
encoded_number = np.arange(len(classes))
classes_df = pd.DataFrame({'Species':classes})
classes_df.head()
```

## Extract the values from train set by stratifying them and dividing it into 80:20 ratio

## Applying all 4 (Randomforest, SVM, NaiveBayes, DecisionTrees) classifiers to find the best one

```python
# Creating the classifiers class
class Classifier:
    def __init__(self,model):
        self.model = model
        self.pred = None
        self.accuracy = None
        self.loss = None
    
    def fit(self,x_train,y_train):
        self.model.fit(x_train,y_train)
    def predict(self,test):
        self.pred = self.model.predict(test)
        return self.pred
    
    def accuracy_score(self,actual):
        self.accuracy = (metrics.accuracy_score(actual,self.pred))*100
        print(self.accuracy)
```

## Applying DecisionTree

```python
DT = Classifier(DecisionTreeClassifier(random_state=50))
```

```python
DT.fit(X_train,Y_train)
```

```python
DT.predict(X_test)
```

```python
DT.accuracy_score(Y_test)
```

### Using Decision Tree Accuracy is 69.7 %

## Applying RandomForest classifier

```python
RFC.accuracy_score(Y_test)
```
