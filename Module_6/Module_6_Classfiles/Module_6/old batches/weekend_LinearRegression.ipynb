```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load the data
bos1 = pd.read_csv('BostonHousing.csv')
bos1.head()
```

```python
# Input = > Independent variable
# outout = > Dependent variable
X = bos1.iloc[:,0:13]
Y = bos1['medv']
```

```python
# split the data into training and testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.30)
```

### Multiple linear regression as we have more than 1 independent variables
#### Y = b1X1+b2X2 + b3X3 +.....b13X13

### simple linear regression: means only one input variable
 y = mX + c

```python
# algorithm
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()

# training
lr_model.fit(x_train, y_train)

# testing
print(x_test)
pred_y = lr_model.predict(x_test)
# known output is y_test and the predicted output is pred_y (i.e algo O/P)
```

```python
data = {'predicted_op':pred_y, 'known_op':y_test}
df = pd.DataFrame(data)
print(df)
```

```python
# Y = mX1+bx2...+ x13+ c
print(lr_model.coef_)
print(lr_model.intercept_)
```
