```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Load the data

```python
boston_data = pd.read_csv("BostonHousing.csv")
boston_data.head()
```

```python
# Input: Independent variable
# Output: Dependent variable
Y = boston_data['medv']
X = boston_data.iloc[:,0:13]
X
```

```python
# let's split the data in training and testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,
                                                    Y,
                                                    test_size=0.3,
                                                    random_state=5)
y_test
```

```python
# using the model
from sklearn.linear_model import LinearRegression

# instantiate the model
lr_model = LinearRegression()
#train the model
lr_model.fit(x_train, y_train)

# testing or prediction
y_pred = lr_model.predict(x_test)

# Compare the house price predicted vs known test value
data = {'predicted O/P': y_pred, 'known(testing) O/P':y_test}
pd.DataFrame(data)
```
