### 5. Compute – Train and Test the data and answer multiple questions -- What is the use of random_state=85? 

```python
corr = df.corr()
sns.heatmap(corr,annot=True)
```

```python
X = df.iloc[:,3:7]
Y = df['Yearly_Amount_Spent']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=85)

## Use of random state is to generate same kind of random selection of data for training and testing for consistent performance
```

### 6. Compute – Predict the data and do a scatter plot. Check if actual and predicted data match?

```python
model = LinearRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
plt.scatter(Y_test,Y_pred)
plt.xlabel('Test Output')
plt.ylabel('Predicted Output')
plt.title('Test Vs Predicted output')
plt.show()
```

### 7. What is the value of Root Mean Squared Error? 

```python
model.coef_
```

```python
mse = metrics.mean_squared_error(Y_test,Y_pred)
mse
```

```python
accuracy = model.score(X_test,Y_test)
print(f"Accuracy is {(accuracy*100).round(2)} %")
```
