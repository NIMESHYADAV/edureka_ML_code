## 2. Data Manipulation: 
### a. Create a new column -’total_benefitted’ that is a sum of inmates benefitted through all modes.

```python
prisoners['total_benefitted'] = prisoners[['No. of Inmates benefitted by Elementary Education',
                                           'No. of Inmates benefitted by Adult Education',
                                           'No. of Inmates benefitted by Higher Education',
                                           'No. of Inmates benefitted by Computer Course']].sum(axis=1)
```

```python
prisoners.head()
```

### b. Create a new row - “totals” that is the sum of all inmates benefitted through each mode across all states. 

```python
prisoners.loc['total']= prisoners[['No. of Inmates benefitted by Elementary Education',
                                           'No. of Inmates benefitted by Adult Education',
                                           'No. of Inmates benefitted by Higher Education',
                                           'No. of Inmates benefitted by Computer Course']].sum()
prisoners
```

```python
sums = prisoners[['No. of Inmates benefitted by Elementary Education',
                                           'No. of Inmates benefitted by Adult Education',
                                           'No. of Inmates benefitted by Higher Education',
                                           'No. of Inmates benefitted by Computer Course']].sum()
print(sums)
```

```python
values = pd.Series(sums.values,dtype='int32')
labels = list(sums.index)
```

## 3. Plotting: 
### a. Make a bar plot with each state name on the x -axis and their total benefitted inmates as their bar heights. Which state has the maximum number of beneficiaries?

```python
plt.figure(figsize=(10,6))
prisoners = prisoners.fillna(0)
prisoners = prisoners.astype('str')
prisoners['total_benefitted'] = prisoners['total_benefitted'].astype('float32')
prisoners.sort_values(by='total_benefitted',inplace=True)
plt.bar(prisoners['STATE/UT'],prisoners['total_benefitted'])
plt.xticks(rotation = 90)
plt.xlabel("State/UT")
plt.ylabel("Total Benefitted inmates")
plt.show()
''' As per the graph -- Andhra Pradesh has the maximum number of beneficiaries '''
```

### b. Make a pie chart that depicts the ratio among different modes of benefits. 

```python
column = sums.index
plt.pie(values,labels=column,autopct='%1.1f%%')
plt.show()
```
