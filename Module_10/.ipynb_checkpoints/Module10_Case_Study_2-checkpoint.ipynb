```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
from sklearn.cluster import KMeans
%matplotlib inline
```

## 1.Open and display the image “dog.jpeg”. Convert the image in to numpy array, so that i can be   used in further processing.  [Hint: Use PIL module from python] 

```python
def load_image(img_name):
    im = Image.open(img_name)
    return im
```

```python
im = load_image('dogs.jpeg')
im
```

### Function to convert an Image to Numpy Array

```python
def img_to_array(im):
    data = np.asarray(im)
    return data
```

```python
dog = img_to_array(im)
dog.shape
```

## 2. Find out the dimensions of the image and convert it in to a two-dimensional array. 

```python
width,height,channel = dog.shape
```

```python
dog_data_2d = dog.reshape(width*height,channel)
```

```python
dog_data_2d.shape
```

## 3.Use kmeans clustering with k set to 3 and cluster the image.      [Hint: Refer to k-means module of scikit learn] 

```python
df = pd.DataFrame(dog_data_2d,columns = ['A','B','C'])
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
plt.scatter(df.A,df.B,df.C,c='red')
plt.show()
```

```python
model = KMeans(n_clusters=3)
model.fit(dog_data_2d)
```

```python
centroids =model.cluster_centers_
centroids
```

```python
cluster_label = model.labels_
cluster_label.shape
```

## 4. Predict the cluster label of every pixel in the image and plot it back as an image. 

```python
image_data = centroids[cluster_label].reshape(width,height,channel)
img_support = np.array(image_data,np.int64)
```

```python
plt.figure(figsize = (12,8))
#cluster_centers = cluster_centers.astype(np.uint8)
plt.imshow(img_support)
plt.show()
```
