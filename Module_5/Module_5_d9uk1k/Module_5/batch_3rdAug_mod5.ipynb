```python
# Iterating dataframe
# 1. iteritems()
df = pd.DataFrame(np.random.randn(5,4),  # 5 rows and 4 columns
                 columns=['col1', 'col2', 'col3', 'col4'])
df
```

```python
for col_name, col_value in df.iteritems():
    print("column name is:",col_name)
    print(col_value)
    print("#"*10)
```

```python
# iterrows() iterates over each row as key, value pair
for row_index, row_value in df.iterrows():
    print("Row index:", row_index)
    print(row_value)
    print("####")
```

```python
df
```

```python
# itertuples() returns an iterator yielding a named tuple for each row
for row in df.itertuples():
    print(row.col1)
    print(row.col2)
    print(row.col3)
    print(row.col4)
    print("#"*10)
```

# GroupBy Operations

```python
import pandas as pd

world_cup={'Team':['West Indies','West Indies','India','Australia','Pakistan','Sri Lanka','Australia','Australia','Australia','India','Australia'],
           'Rank':[7,7,2,1,6,4,1,1,1,2,1],
           'Year':[1975,1979,1983,1987,1992,1996,1999,2003,2007,2011,2015]}
df=pd.DataFrame(world_cup)
print(df)
```

```python
df.groupby('Team')
```

```python
df.groupby('Team').groups
```

```python
# how to access a group
# a single group can also be selected using get_group()
grouped = df.groupby('Team')
grouped.get_group('India')
```

```python
grouped.get_group('Australia')
```

```python
# grouping based on more than 1 column
df.groupby(['Team', 'Rank']).groups
```

```python
grouped = df.groupby(['Team', 'Rank'])
grouped.get_group(('Australia', 1))
```

# Concatenation

It is a process of combining two or more data structures


```python
# method 1
import pandas as pd

world_champions={'Team':['India','Australia','West Indies','Pakistan','Sri Lanka'],
           'ICC_rank':[2,3,7,8,4],
           'World_champions_Year':[2011,2015,1979,1992,1996],
           'Points':[874,787,753,673,855]}

chokers={'Team':['South Africa','New Zealand','Zimbabwe'],
           'ICC_rank':[1,5,9],
           'Points':[895,764,656]}
df1 = pd.DataFrame(world_champions)
df2 = pd.DataFrame(chokers)
print("DF1: World champions")
df1
```

```python
print("DF2: Chokers")
df2
```

```python
# concatenation
df = pd.concat([df1, df2], sort=False)
df
```

```python
# method 2 to do concatenation
import pandas as pd

world_champions={'Team':['India','Australia','West Indies','Pakistan','Sri Lanka'],
           'ICC_rank':[2,3,7,8,4],
           'World_champions_Year':[2011,2015,1979,1992,1996],
           'Points':[874,787,753,673,855]}

chokers={'Team':['South Africa','New Zealand','Zimbabwe'],
           'ICC_rank':[1,5,9],
           'Points':[895,764,656]}
df1 = pd.DataFrame(world_champions)
df2 = pd.DataFrame(chokers)
df = df1.append(df2, sort=False)
df
```

```python
# concatenating using axis=1 i.e adding columns
world_champions={'Team':['India','Australia','West Indies','Pakistan','Sri Lanka'],
           'ICC_rank':[2,3,7,8,4],
           'World_champions_Year':[2011,2015,1979,1992,1996],
           'Points':[874,787,753,673,855]}

chokers={'Team':['South Africa','New Zealand','Zimbabwe'],
           'ICC_rank':[1,5,9],
           'Points':[895,764,656]}
df1 = pd.DataFrame(world_champions)
df2 = pd.DataFrame(chokers)
df = pd.concat([df1, df2], axis=1)
df
```

# Merging and Joining
Merging is the pandas operation that peforms database joins on object

```python
import pandas as pd

world_champions={'Team':['India','Australia','West Indies','Pakistan','Sri Lanka'],
           'ICC_rank':[2,3,7,8,4],
           'World_champions_Year':[2011,2015,1979,1992,1996],
           'Points':[874,787,753,673,855]}

chokers={'Team':['India','New Zealand','Zimbabwe'],
           'ICC_rank':[1,5,9],
           'Points':[777777,764,656]}
df1 = pd.DataFrame(world_champions)
df2 = pd.DataFrame(chokers)
df1
```

```python
df2
```

```python
# merge dataframes on a key
# inner join
df = pd.merge(df1, df2, on='Team')
df
```

```python
# Left Join
world_champions={'Team':['India','Australia','West Indies','Pakistan','Sri Lanka'],
           'ICC_rank':[2,3,7,8,4],
           'World_champions_Year':[2011,2015,1979,1992,1996],
           'Points':[874,787,753,673,855]}

chokers={'Team':['India','New Zealand','Zimbabwe'],
           'ICC_rank':[1,5,9],
           'Points':[777777,764,656]}
df1 = pd.DataFrame(world_champions)
df2 = pd.DataFrame(chokers)
df1
```

```python
df2
```

```python
df = pd.merge(df1, df2, on='Team', how='left')
df
```

```python
df = pd.merge(df1, df2, on='Team', how='right')
df
```

```python
# full outer join
# outer join: combines into one like full union.
world_champions={'Team':['India','Australia','West Indies','Pakistan','Sri Lanka'],
           'ICC_rank':[2,3,7,8,4],
           'World_champions_Year':[2011,2015,1979,1992,1996],
           'Points':[874,787,753,673,855]}

chokers={'Team':['India','New Zealand','Zimbabwe'],
           'ICC_rank':[1,5,9],
           'Points':[777777,764,656]}
df1 = pd.DataFrame(world_champions)
df2 = pd.DataFrame(chokers)
df1
```

```python
df2
```

```python
pd.merge(df1, df2, on='Team', how='outer')
```

## Case Study - AllCountries

```python
import pandas as pd
```

```python
dataset = pd.read_csv('AllCountries.csv')
dataset.head()
```

```python
dataset.shape
```

```python
# datatypes of columns
dataset.dtypes
```

```python
dataset.info()
```

```python
# staistical summary
dataset.describe()
```

## 1. find countries with size greater than 2000units

```python
# Method 1
bigger_countries = dataset[dataset['LandArea'] > 2000]
bigger_countries
```

```python
# Method 2
for row in dataset.itertuples():
    if row.LandArea > 2000:
        print(row.Country, row.LandArea)
        print("####")
```

```python
# sorting of a dataframe to get the countries sorted based on LandArea
bigger_countries.sort_values(by='LandArea', ascending=False).head()
```

## 2. Is there any correlation between GDP and Birth rate ?

```python
import matplotlib.pyplot as plt
GDP = dataset['GDP']
BirthRate = dataset['BirthRate']
plt.scatter(x=GDP,y=BirthRate)
plt.xlabel("GDP")
plt.ylabel("BirthRate")
plt.show()
```

```python
plt.figure(figsize=(5,5))  # to control the plot size
plt.scatter(x=GDP,y=BirthRate)
plt.xlabel("GDP")
plt.ylabel("BirthRate")
plt.xlim(0, 20000)
plt.show()
```

## 3. Compare the GDP of ten richest countries

```python
relevant_data = dataset.loc[:,['Country', 'GDP', 'BirthRate']]
top_10 = relevant_data.sort_values(by='GDP', ascending=False).head(10)
top_10
```

```python
# bar chart
plt.bar(top_10['Country'],top_10['GDP'])
plt.xlabel('Countries')
plt.ylabel('GDP')
plt.xticks(rotation=90)
plt.show()
```

```python
# pie chart
plt.pie(top_10['GDP'],
        labels=top_10['Country'],
        autopct="%.2f%%")
plt.show()
```
