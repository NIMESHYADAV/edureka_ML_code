```python
# Iterating over a DataFrame - iteritems()
data = np.random.randn(5,4)
df = pd.DataFrame(data, columns=['co11', 'col2', 'col3','col4'])
df
```

```python
#1. iteritems(): Iterate over the DF on the basis of column
for key, value in df.iteritems():
    print("Key is Column name: ", key)
    print(value)
    print("#"*10)
```

```python
#2. iterrows() :iterate over dataframe , row wise
for key, value in df.iterrows():
    print("Key id Row index: ", key)
    print(value)
    print("#"*20)
```

```python
#3. itertuples() : returns an iterator having named tuple for each row
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.itertuples.html
data = np.random.randn(5,4)
df = pd.DataFrame(data, columns=['col1', 'col2', 'col3','col4'])
print("****Original DataFrame*****")
print(df)
print("$"*30)
for row in df.itertuples(name='Edureka'): #name='Edureka' is optional
    print(row)  #named tuple
    print(row.col1)
    print("#"*20)
    
```

### GroupBy

```python
world_cup = {'Team':['West Indies', 'West Indies', 'India', 'Australia',
                     'Pakistan', 'Sri Lanka', 'Australia', 'Australia','Australia',
                     'India','Australia'],
            'Rank':[7,7,2,1,6,4,1,1,1,2,1],
            'Year':[1975, 1979, 1983, 1987, 1992, 1996, 1999,2003, 2007,2011, 2015]}
df = pd.DataFrame(world_cup)
df
```

```python
# groupBy()
df.groupby('Team').groups
```

```python
df.groupby(['Team', 'Rank']).groups
```

```python
# we can iterate over the groups
grouped = df.groupby('Team')
for name, group in grouped:
    print("Group Name/key: ",name)
    print(group)
    print("#"*20)
```

```python
# to access a specific group
grouped = df.groupby('Team')
grouped.get_group('India')
```

## Concatenation
- process of combining two or more data structures

```python
import pandas as pd

world_champions={'Team':['India','Australia','West Indies','Pakistan','Sri Lanka'],
           'ICC_rank':[2,3,7,8,4],
           'World_champions_Year':[2011,2015,1979,1992,1996],
           'Points':[874,787,753,673,855]}

chokers={'Team':['South Africa','New Zealand','Zimbabwe'],
           'ICC_rank':[1,5,9],
           'Points':[895,764,656]}
df1 = pd.DataFrame(world_champions)
df2 = pd.DataFrame(chokers)
print("DF1: World champions")
print(df1)
print("#"*20)
print("DF2: Chokers")
print(df2)
```

```python
# concat()
pd.concat([df1,df2], sort=True)
```

```python
# append()
world_champions={'Team':['India','Australia','West Indies','Pakistan','Sri Lanka'],
           'ICC_rank':[2,3,7,8,4],
           'World_champions_Year':[2011,2015,1979,1992,1996],
           'Points':[874,787,753,673,855]}

chokers={'Team':['South Africa','New Zealand','Zimbabwe'],
           'ICC_rank':[1,5,9],
           'Points':[895,764,656]}
df1 = pd.DataFrame(world_champions)
df2 = pd.DataFrame(chokers)
print("DF1: World champions")
print(df1)
print("#"*20)
print("DF2: Chokers")
print(df2)
print("******** Append output ********")
print(df1.append(df2, sort=True))
```

```python
# axis=1
world_champions={'Team':['India','Australia','West Indies','Pakistan','Sri Lanka'],
           'ICC_rank':[2,3,7,8,4],
           'World_champions_Year':[2011,2015,1979,1992,1996],
           'Points':[874,787,753,673,855]}

chokers={'Team':['South Africa','New Zealand','Zimbabwe'],
           'ICC_rank':[1,5,9],
           'Points':[895,764,656]}
df1 = pd.DataFrame(world_champions)
df2 = pd.DataFrame(chokers)
print("DF1: World champions")
print(df1)
print("#"*20)
print("DF2: Chokers")
print(df2)
print("******** Concatenation Column wise output ********")
print(pd.concat([df1,df2],axis=1))
```

## Merging and Joining
- merging performs Database Joins

```python
champion_stats={'Team':['India','Australia','West Indies','Pakistan','Sri Lanka'],
           'ICC_rank':[2,3,7,8,4],
           'World_champions_Year':[2011,2015,1979,1992,1996],
           'Points':[874,787,753,673,855]}

match_stats={'Team':['India','Australia','West Indies','Pakistan','Sri Lanka'],
             'World_cup_played':[11,10,11,9,8],
             'ODIs_played':[733,988,712,679,662]}

df1 = pd.DataFrame(champion_stats)
df2 = pd.DataFrame(match_stats)
print("****** DataFrame:champion_stats******")
print(df1)
print("****** DataFrame:match_stats******")
print(df2)
```

```python
# merge()
pd.merge(df1, df2, on='Team')
```

```python
# left join merges two dataframes based on the keys of the left dataframe
world_champions={'Team':['India','Australia','West Indies','Pakistan','Sri Lanka'],
           'ICC_rank':[2,3,7,8,4],
           'World_champions_Year':[2011,2015,1979,1992,1996],
           'Points':[874,787,753,673,855]}

chokers={'Team':['India','New Zealand','Zimbabwe'],
           'ICC_rank':[2,5,9],
           'Points':[874,764,656]}
df1 = pd.DataFrame(world_champions)
df2 = pd.DataFrame(chokers)
print("DF1: World champions")
print(df1)
print("#"*20)
print("DF2: Chokers")
print(df2)
print(" *******Left Join Merge*********")
print(pd.merge(df1, df2, on='Team', how='left'))
```

```python
# right join merges two dataframes based on the keys of the right dataframe
world_champions={'Team':['India','Australia','West Indies','Pakistan','Sri Lanka'],
           'ICC_rank':[2,3,7,8,4],
           'World_champions_Year':[2011,2015,1979,1992,1996],
           'Points':[874,787,753,673,855]}

chokers={'Team':['India','New Zealand','Zimbabwe'],
           'ICC_rank':[2,5,9],
           'Points':[874,764,656]}
df1 = pd.DataFrame(world_champions)
df2 = pd.DataFrame(chokers)
print("DF1: World champions")
print(df1)
print("#"*20)
print("DF2: Chokers")
print(df2)
print(" *******Right Join Merge*********")
print(pd.merge(df1, df2, on='Team', how='right'))
```

```python
# outer join merges two dataframes based on the full union
world_champions={'Team':['India','Australia','West Indies','Pakistan','Sri Lanka'],
           'ICC_rank':[2,3,7,8,4],
           'World_champions_Year':[2011,2015,1979,1992,1996],
           'Points':[874,787,753,673,855]}

chokers={'Team':['India','New Zealand','Zimbabwe'],
           'ICC_rank':[2,5,9],
           'Points':[874,764,656]}
df1 = pd.DataFrame(world_champions)
df2 = pd.DataFrame(chokers)
print("DF1: World champions")
print(df1)
print("#"*20)
print("DF2: Chokers")
print(df2)
print(" *******outer Join Merge*********")
print(pd.merge(df1, df2, on='Team', how='outer'))
```

```python
# inner join returns intersection of two Dataframes.
world_champions={'Team':['India','Australia','West Indies','Pakistan','Sri Lanka'],
           'ICC_rank':[2,3,7,8,4],
           'World_champions_Year':[2011,2015,1979,1992,1996],
           'Points':[874,787,753,673,855]}

chokers={'Team':['India','New Zealand','Zimbabwe'],
           'ICC_rank':[2,5,9],
           'Points':[874,764,656]}
df1 = pd.DataFrame(world_champions)
df2 = pd.DataFrame(chokers)
print("DF1: World champions")
print(df1)
print("#"*20)
print("DF2: Chokers")
print(df2)
print(" *******Inner Join Merge*********")
print(pd.merge(df1, df2, on='Team', how='inner'))
```

```python
#1. Find countries having LandArea greater than 2000 units.
# method 1: iterating over the data
selected_data = data_df.loc[:,['Country', 'LandArea']]
for named_tuple_obj in selected_data.itertuples():
    if named_tuple_obj.LandArea > 2000:
        print(named_tuple_obj.Country)
```

```python
# method 2: applying the filter directly to the complete Dataframe.
# giving one filter
data_df[data_df['LandArea'] > 2000]
```

```python
#giving two filters
data_df[(data_df['LandArea'] > 2000) & (data_df['Population']>50)]
```

```python
# 2. See if there is a correlation between GDP per capita and
# ..Birth rate of countries ??
import matplotlib.pyplot as plt

selected_data = data_df.loc[:,['Country', 'GDP', 'BirthRate']]
x = selected_data['GDP']
y = selected_data['BirthRate']
plt.scatter(x,y)
plt.xlabel('GDP')
plt.ylabel('Birth Rate')
plt.show()

```

```python
# 2.1
plt.scatter(x,y)
plt.xlabel('GDP')
plt.ylabel('Birth Rate')
plt.xlim(0,20000)
plt.show()
```

```python
#3.compare the GDP of ten richest countries
import matplotlib.pyplot as plt

selected_data = data_df.loc[:,['Country', 'GDP']]
sorted_data = selected_data.sort_values(by='GDP', ascending=False)
# sort_index is deprecated
#sorted_data = selected_data.sort_index(by='GDP', ascending=False)
top10_countries_gdp = sorted_data.iloc[:10]
print(top10_countries_gdp)

plt.pie(top10_countries_gdp['GDP'],
        labels=top10_countries_gdp['Country'],
        autopct="%1.1f%%"
       )

plt.show()

```
