```python
# Iterating dataframe
# 1. iteritems()
df = pd.DataFrame(np.random.randn(5,4),  # 5 rows and 4 columns
                 columns=['col1', 'col2', 'col3', 'col4'])
df
```

```python
for col_name, col_value in df.iteritems():
    print("column name is:",col_name)
    print(col_value)
    print("#"*10)
```

```python
# iterrows() iterates over each row as key, value pair
for row_index, row_value in df.iterrows():
    print("Row index:", row_index)
    print(row_value)
    print("####")
```

```python
# itertuples() returns an iterator yielding a named tuple for each row
for row in df.itertuples():
    print(row.col1)
    print(row.col2)
    print(row.col3)
    print(row.col4)
    print("#"*10)
```

# GroupBy Operations

```python
import pandas as pd

world_cup={'Team':['West Indies','West Indies','India','Australia','Pakistan','Sri Lanka','Australia','Australia','Australia','India','Australia'],
           'Rank':[7,7,2,1,6,4,1,1,1,2,1],
           'Year':[1975,1979,1983,1987,1992,1996,1999,2003,2007,2011,2015]}
df=pd.DataFrame(world_cup)
print(df)
```

```python
df.groupby('Team').groups  # .groups is used to get the values in dict
```

```python
df.groupby(['Team', 'Rank']).groups
```

```python
df.groupby(['Team', 'Rank'])
```

```python
# how to iterate over a group object
for name, group in df.groupby('Team'):
    print(name)
    #print(group)
```

```python
# a single group can also be selected using get_group()
grouped = df.groupby('Team')
grouped.get_group('India')
```

# Concatenation

It is a process of combining two or more data structures


```python
# method 1
import pandas as pd

world_champions={'Team':['India','Australia','West Indies','Pakistan','Sri Lanka'],
           'ICC_rank':[2,3,7,8,4],
           'World_champions_Year':[2011,2015,1979,1992,1996],
           'Points':[874,787,753,673,855]}

chokers={'Team':['South Africa','New Zealand','Zimbabwe'],
           'ICC_rank':[1,5,9],
           'Points':[895,764,656]}
df1 = pd.DataFrame(world_champions)
df2 = pd.DataFrame(chokers)
print("DF1: World champions")
print(df1)
print("#"*20)
print("DF2: Chokers")
print(df2)
```

```python
# concatenation
df = pd.concat([df1, df2], sort=False)
df
```

```python
df.loc[0] # returns entries with index(key) 0
```

```python
df.iloc[0]  # return value at that position
```

```python
df.iloc[1]
```

```python
# method 2 to do concatenation
import pandas as pd

world_champions={'Team':['India','Australia','West Indies','Pakistan','Sri Lanka'],
           'ICC_rank':[2,3,7,8,4],
           'World_champions_Year':[2011,2015,1979,1992,1996],
           'Points':[874,787,753,673,855]}

chokers={'Team':['South Africa','New Zealand','Zimbabwe'],
           'ICC_rank':[1,5,9],
           'Points':[895,764,656]}
df1 = pd.DataFrame(world_champions)
df2 = pd.DataFrame(chokers)
df = df1.append(df2, sort=False)
df
```

```python
# concatenating using axis=1 i.e adding columns
world_champions={'Team':['India','Australia','West Indies','Pakistan','Sri Lanka'],
           'ICC_rank':[2,3,7,8,4],
           'World_champions_Year':[2011,2015,1979,1992,1996],
           'Points':[874,787,753,673,855]}

chokers={'Team':['South Africa','New Zealand','Zimbabwe'],
           'ICC_rank':[1,5,9],
           'Points':[895,764,656]}
df1 = pd.DataFrame(world_champions)
df2 = pd.DataFrame(chokers)
df = pd.concat([df1, df2], axis=1)
df
```

# Merging and Joining
Merging is the pandas operation that peforms database joins on object

```python
import pandas as pd

champion_stats={'Team':['India','Australia','West Indies','Pakistan','Sri Lanka'],
           'ICC_rank':[2,3,7,8,4],
           'World_champions_Year':[2011,2015,1979,1992,1996],
           'Points':[874,787,753,673,855]}

match_stats={'Team':['India','Australia','West Indies','Pakistan','Sri Lanka'],
             'World_cup_played':[11,10,11,9,8],
             'ODIs_played':[733,988,712,679,662]}


df1=pd.DataFrame(champion_stats)
df2=pd.DataFrame(match_stats)
print(df1)
print(df2)
```

```python
# merge dataframes on a key
df = pd.merge(df1, df2, on='Team')
df
```

```python
# access column with the index
df.iloc[:,1]
```

```python
# access column via name
df['ICC_rank']
```

```python
# Left Join
world_champions={'Team':['India','Australia','West Indies','Pakistan','Sri Lanka'],
           'ICC_rank':[2,3,7,8,4],
           'World_champions_Year':[2011,2015,1979,1992,1996],
           'Points':[874,787,753,673,855]}

chokers={'Team':['India','New Zealand','Zimbabwe'],
           'ICC_rank':[1,5,9],
           'Points':[777777,764,656]}
df1 = pd.DataFrame(world_champions)
df2 = pd.DataFrame(chokers)
print(df1)
print("#######")
print(df2)
```

```python
pd.merge(df1, df2, on='Team', how='left')
```

```python
world_champions={'Team':['India','Australia','West Indies','Pakistan','Sri Lanka'],
           'ICC_rank':[2,3,7,8,4],
           'World_champions_Year':[2011,2015,1979,1992,1996],
           'Points':[874,787,753,673,855]}

chokers={'Team':['India','New Zealand','Zimbabwe'],
           'ICC_rank':[1,5,9],
           'Points':[777777,764,656]}
df1 = pd.DataFrame(world_champions)
df2 = pd.DataFrame(chokers)
print(df1)
print("#######")
print(df2)
```

```python
pd.merge(df1, df2, on='Team', how='right')
```

```python
# outer join: combines into one like full union.
world_champions={'Team':['India','Australia','West Indies','Pakistan','Sri Lanka'],
           'ICC_rank':[2,3,7,8,4],
           'World_champions_Year':[2011,2015,1979,1992,1996],
           'Points':[874,787,753,673,855]}

chokers={'Team':['India','New Zealand','Zimbabwe'],
           'ICC_rank':[1,5,9],
           'Points':[777777,764,656]}
df1 = pd.DataFrame(world_champions)
df2 = pd.DataFrame(chokers)
pd.merge(df1, df2, on='Team', how='outer')
```

```python
# inner join: merges two objects based on the intersection of the columns of..
# ...both the objects.

world_champions={'Team':['India','Australia','West Indies','Pakistan','Sri Lanka'],
           'ICC_rank':[2,3,7,8,4],
           'World_champions_Year':[2011,2015,1979,1992,1996],
           'Points':[874,787,753,673,855]}

chokers={'Team':['India','New Zealand','Zimbabwe'],
           'ICC_rank':[1,5,9],
           'Points':[777777,764,656]}
df1 = pd.DataFrame(world_champions)
df2 = pd.DataFrame(chokers)
pd.merge(df1, df2, on='Team', how='inner')
```
